```python
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 4*np.pi, 0.01)
m = 0.5

cos = np.cos(x)
plt.plot(x, cos)

x = np.arange(0, np.pi-m, 0.01)
plt.plot(x, np.cos(x+m), lw=3)

x = np.arange(np.pi-m, 4*np.pi, 0.01)
plt.plot(x, -np.cos(x+m)-2, lw=3)
plt.legend(["Cos(x)", "cos(x+m)", "-cos(x+m)-2"])
plt.grid()
```

```python
x = np.arange(0, 4*np.pi, 0.01)
m = 0.5

cos = np.cos(x)
plt.plot(x, cos)

x = np.arange(0, np.pi-m, 0.01)
plt.plot(x, np.cos(x+m), lw=3)

x = np.arange(np.pi-m, 4*np.pi, 0.01)
y = np.cos(x) - np.sin(np.pi-m)*m
plt.plot(x, y, lw=3)
plt.legend(["Cos(x)", "cos(x+m)", "-cos(x+m)-2"])
plt.grid()
```
