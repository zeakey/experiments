```python
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 4*np.pi, 0.01)
m = 0.5

cos = np.cos(x)
plt.plot(x, cos)

x = np.arange(0, np.pi-m, 0.01)
plt.plot(x, np.cos(x+m), lw=3)

x = np.arange(np.pi-m, 4*np.pi, 0.01)
plt.plot(x, -np.cos(x+m)-2, lw=3)
plt.legend(["Cos(x)", "cos(x+m)", "-cos(x+m)-2"])
plt.grid()
```

```python
x = np.arange(0, 4*np.pi, 0.01)
m = 0.5

cos = np.cos(x)
plt.plot(x, cos)

x = np.arange(0, np.pi-m, 0.01)
plt.plot(x, np.cos(x+m), lw=3)

x = np.arange(np.pi-m, 4*np.pi, 0.01)
y = np.cos(x) - np.sin(np.pi-m)*m
plt.plot(x, y, lw=3)
plt.legend(["Cos(x)", "cos(x+m)", "-cos(x+m)-2"])
plt.grid()
```

```python
x = np.arange(0, np.pi, 0.1)
N = 10572

s = 8
p0 = np.exp(s*np.cos(x)) / (np.exp(s*np.cos(x)) + N - 1)
loss0 = -np.log(p0)

s = 16
p1 = np.exp(s*np.cos(x)) / (np.exp(s*np.cos(x)) + N - 1)
loss1 = -np.log(p1)

s = 32
p2 = np.exp(s*np.cos(x)) / (np.exp(s*np.cos(x)) + N - 1)
loss2 = -np.log(p2)

s = 64
p3 = np.exp(s*np.cos(x)) / (np.exp(s*np.cos(x)) + N - 1)
loss3 = -np.log(p3)

plt.grid()
plt.plot(x, loss0)
plt.plot(x, loss1)
plt.plot(x, loss2)
plt.plot(x, loss3)
plt.xlabel("$\\theta$", fontsize=20)
plt.ylabel("Loss", fontsize=20)
plt.legend(["s=8", "s=16", "s=32", "s=64"])
```
